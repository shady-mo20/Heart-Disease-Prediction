```python
heart_data.rename(columns={
    'Heart Disease': 'Heart_Disease',
    'Chest pain type': 'Chest_Pain_Type',
    'FBS over 120': 'FBS_over_120',
    'Number of vessels fluro': 'Num_Vessels_Fluoro'
}, inplace=True)
heart_data['Heart_Disease'] = heart_data['Heart_Disease'].map({'Presence': 1, 'Absence': 0})

def handle_missing_values(df):
    numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
    df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
    for col in categorical_cols:
        df[col] = df[col].fillna(df[col].mode()[0])
    return df

heart_data = handle_missing_values(heart_data)

def create_comprehensive_report(df, report_filename='Comprehensive_Heart_Disease_Report.xlsx'):
    corr_heatmap_path = 'corr_heatmap.png'
    age_boxplot_path = 'age_boxplot.png'
    
    numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    corr_matrix = df[numeric_cols].corr()
    
    plt.figure(figsize=(12, 10))
    sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
    plt.title('Correlation Matrix Heatmap')
    plt.tight_layout()
    plt.savefig(corr_heatmap_path)
    plt.close()

    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Sex', y='Age', data=df)
    plt.title('Age Distribution by Sex')
    plt.savefig(age_boxplot_path)
    plt.close()
    
    with pd.ExcelWriter(report_filename, engine='xlsxwriter') as writer:
        df.to_excel(writer, sheet_name='Raw Data', index=False)
        descriptive_stats = df.describe().transpose()
        descriptive_stats.to_excel(writer, sheet_name='Descriptive Statistics')
        
        categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
        if categorical_cols:
            categorical_summary = df[categorical_cols].describe().transpose()
            categorical_summary.to_excel(writer, sheet_name='Categorical Summary')
        
        corr_matrix.to_excel(writer, sheet_name='Correlation Matrix')
        
        t_statistic, p_value = stats.ttest_ind(
            df[df['Sex'] == 0]['Age'],
            df[df['Sex'] == 1]['Age']
        )
        t_test_results = pd.DataFrame({
            'Test': ['Independent T-Test'],
            'Group 1': ['Sex 0'],
            'Group 2': ['Sex 1'],
            'T-statistic': [t_statistic],
            'P-value': [p_value]
        })
        t_test_results.to_excel(writer, sheet_name='T-Test Results', index=False)
        
        workbook = writer.book
        worksheet = workbook.add_worksheet('Charts')
        writer.sheets['Charts'] = worksheet

        heart_disease_counts = df['Heart_Disease'].value_counts()
        chart1 = workbook.add_chart({'type': 'column'})
        chart1.add_series({
            'name': 'Heart Disease Count',
            'categories': ['Raw Data', 1, df.columns.get_loc('Heart_Disease'), len(df), df.columns.get_loc('Heart_Disease')],
            'values':     ['Raw Data', 1, df.columns.get_loc('Heart_Disease'), len(df), df.columns.get_loc('Heart_Disease')],
            'fill': {'color': '#5ABA10'}
        })
        chart1.set_title({'name': 'Heart Disease Distribution'})
        chart1.set_x_axis({'name': 'Heart Disease (0: Absence, 1: Presence)'})
        chart1.set_y_axis({'name': 'Count'})
        worksheet.insert_chart('B2', chart1)
        
        worksheet.insert_image('B20', corr_heatmap_path, {'x_scale': 0.5, 'y_scale': 0.5})
        worksheet.insert_image('B40', age_boxplot_path, {'x_scale': 0.5, 'y_scale': 0.5})
    
    os.remove(corr_heatmap_path)
    os.remove(age_boxplot_path)
    
    print(f"Comprehensive report created successfully in: {report_filename}")

create_comprehensive_report(heart_data)
```
